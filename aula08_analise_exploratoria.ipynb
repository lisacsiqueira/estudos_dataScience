```python
# receita total
df["Valor Venda"].sum()
```

```python
#Qual o custo total por venda?
#Custo total =  custo unitario * prd vendido

df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) # criando a coluna
```

```python
df.head()
```

```python
# custo total
# round - arredonda as casas decimais 

round(df["custo"].sum(),2)

```

```python
# agora que temos a receita e custo  -> acha-se o lucro 
df["lucro"] = df["Valor Venda"] - df["custo"] # criou a coluna lucro
```

```python
df.head(1)
```

```python
#lucro total

round(df["lucro"].sum(),2)
```

```python
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
# extraindo apenas days->int
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
# verificando o tipo de tempo_envio

df["Tempo_envio"].dtype
```

```python
# media do tempo_envio

'''
groupby agrupa pelo nome igual que está na coluna 
'''

df.groupby("Marca")["Tempo_envio"].mean()
```
